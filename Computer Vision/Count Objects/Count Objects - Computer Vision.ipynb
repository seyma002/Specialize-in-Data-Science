Object counting is an important research topic in computer vision and is used in many application areas. This project aims to automatically count specific objects in images. The project aims to perform the object counting task using image processing.

```python
import numpy as np
```

```python
import cv2
import matplotlib.pyplot as plt
import cvlib
from cvlib.object_detection import draw_bbox
```

```python
img=cv2.imread('cars.jpg')
plt.imshow(img)
```

```python
img=cv2.imread('cars.jpg')
box,label,count=cvlib.detect_common_objects(img)
output=draw_bbox(img,box,label,count)
plt.figure(figsize=(10,10))
plt.imshow(output)
print("Number of cars in this image are " +str(label.count('car')))
```

```python
img=cv2.imread('giraffe.jpg')
box,label,count=cvlib.detect_common_objects(img)
output=draw_bbox(img,box,label,count)
plt.figure(figsize=(8,8))
plt.imshow(output)
print("Number of giraffes in this image are " +str(label.count('giraffe')))
```

```python
image = cv2.imread("flowers.jpg")
detector =cv2.SimpleBlobDetector_create()
keypoints = detector.detect(image) 
cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS 
blank = np.zeros((1,1)) 
blobs = cv2.drawKeypoints(image, keypoints, blank, (255,0,0),cv2.DRAW_MATCHES_FLAGS_DEFAULT)
num_blobs = len(keypoints)
print(f"Number of flowers detected: {num_blobs}")
cv2.imshow("Blobs", blobs)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
