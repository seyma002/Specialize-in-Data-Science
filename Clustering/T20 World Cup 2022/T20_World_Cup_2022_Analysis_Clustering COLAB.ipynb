## T20 World Cup 2022 Analysis - Clustering

The T20 World Cup 2022 is an international tournament held in the T20 format, one of the most exciting formats of cricket. In this project, we will conduct a clustering analysis to analyze the performances of teams and players and identify their similarities using the data collected throughout the tournament. In this way, it will be possible to better understand the performance characteristics of teams and players.

The dataset to be used for the project includes information on matches played during the T20 World Cup 2022, player statistics, and team performances. statistics to be used to evaluate performance.

Cluster analysis is a machine learning technique used to group data with similar characteristics.

As a result of this project, a deeper understanding of the performances of teams and players in the T20 World Cup 2022 will be gained. With cluster analysis, teams and players with similar performance characteristics will be grouped, which can help in making strategic decisions for future tournaments.

**Information About Columns**

* venue: The place or stadium where the match is played.

* team1: The first team in the match.

* team2: The second team in the match.

* stage: The stage at which the match is played (for example, group stage, quarterfinals, semifinals, final).

* toss winner: The team that wins the toss.

* toss decision: The decision made by the team that wins the toss (for example, choosing the field or choosing to bat).

* first innings score: The total points scored during the first innings.

* first innings wickets: The number of wickets taken during the first innings.

* second innings score: The total points scored during the second innings.

* second innings wickets: The number of wickets taken during the second innings.

* winner: The team that won the match.

* won by: The method by which the match was won (for example, by points, by wickets).

* player of the match: The player who performed best in the match.

* top scorer: The player who scored the most points in the match.
* highest score: The score of the player who scored the highest score in the match.

* best bowler: The bowler who performed best in the match.

* best bowling figure: The best bowling statistic (for example, runs scored and wickets taken).

### Modelling

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn import datasets
from yellowbrick.cluster import KElbowVisualizer
```

```python
df['won by'].value_counts()
```

```python
d={'Runs':0, 'Wickets':1}
```

```python
df['won by']=df['won by'].map(d)
```

```python
x=df[['first innings score','first innings wickets','second innings score','second innings wickets','won by']]
```

```python
model=KMeans(4)
```

```python
model=model.fit(x)
```

```python
tahmin=model.predict(x)
```

```python
x['cluster']=tahmin
```

```python
silhouette_score(x, tahmin)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
wcss=[]
ss=[]
for i in range(2,10):
    model=KMeans(i)
    model=model.fit(x)
    tahmin=model.predict(x)
    ss1=silhouette_score(x, tahmin)
    ss.append(ss1)
    print(ss1)
    wcss.append(model.inertia_)
```
